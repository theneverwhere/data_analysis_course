```python
df = pd.read_csv('beverage_r.csv', sep=';', encoding='utf-8', index_col=0)
df.head()
```

```python
# Разделить таблицу данных так, чтобы в обучающую выборку попало 66% наблюдений, а в тестовую выборку 33% наблюдений
train_df = df.sample(frac=0.66, random_state=42)
test_df = df.drop(train_df.index)
```

```python
# Разделить таблицу данных так, чтобы в обучающую выборку попало 66% наблюдений, а в тестовую выборку 33% наблюдений
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.33, random_state=42)
print('Соотношение размеров обучающей и тестовой выборок в процентах:')
print(f'Обучающая выборка: {len(train_df) / len(df) * 100:.2f}%')
print(f'Тестовая выборка: {len(test_df) / len(df) * 100:.2f}%')
train_df.head()
```

```python
# Разделить таблицу данных так, чтобы в обучающую выборку попало 66% наблюдений,
# а в тестовую выборку 33% наблюдений без использования sklearn и pandas sample.
df = pd.read_csv('beverage_r.csv', sep=';', encoding='utf-8')
df.head()
np.random.seed(42)  # для воспроизводимости
mask = np.random.rand(len(df)) < 0.5
train_df = df[mask]
test_df = df[~mask]
print('Соотношение размеров обучающей и тестовой выборок в процентах:')
print(f'Обучающая выборка: {len(train_df) / len(df) * 100:.2f}%')
print(f'Тестовая выборка: {len(test_df) / len(df) * 100:.2f}%')
train_df.head()
```

```python
# Разделить таблицу данных на четыре одинаковых части c помощью склерн
from sklearn.model_selection import train_test_split
part1, temp = train_test_split(df, test_size=0.75, random_state=42)
part2, temp = train_test_split(temp, test_size=0.6667, random_state=42)
part3, part4 = train_test_split(temp, test_size=0.5, random_state=42)
print(f'Размер части 1: {len(part1)}')
print(f'Размер части 2: {len(part2)}')
print(f'Размер части 3: {len(part3)}')
print(f'Размер части 4: {len(part4)}')
```

```python
# Разделить таблицу данных на две стратифицированные подвыборки.
# По переменным кока-кола и pepsi
from sklearn.model_selection import train_test_split
stratify_cols = df[['SPRITE', 'PEPSI']]
train_df, test_df = train_test_split(df, test_size=0.33, random_state=42, stratify=stratify_cols)
print('Соотношение размеров обучающей и тестовой выборок в процентах:')
print(f'Обучающая выборка: {len(train_df) / len(df) * 100:.2f}%')
print(f'Тестовая выборка: {len(test_df) / len(df) * 100:.2f}%')
```

```python
# доказательство что стратификация сработала
print('Распределение в исходной таблице:')
print(df[['SPRITE', 'PEPSI']].value_counts(normalize=True).sort_index())
print('Распределение в обучающей выборке:')
print(train_df[['SPRITE', 'PEPSI']].value_counts(normalize=True).sort_index())
print('Распределение в тестовой выборке:')
print(test_df[['SPRITE', 'PEPSI']].value_counts(normalize=True).sort_index())
```
